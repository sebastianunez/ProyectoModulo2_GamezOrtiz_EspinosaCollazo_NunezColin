```python
# Cargamos hoja de calculo en un dataframe
file_temp = 'Datos de temperatura.csv'
df_temp = pd.read_csv(file_temp, usecols=[ 'Media'])
# df_temp = pd.read_csv(file_temp, index_col='Fecha', usecols=['Fecha', 'Media'])
df_temp
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#df_temp.plot(figsize=(8,6));
plt.plot(df_temp.index, df_temp.loc[:,'Media'], 'r*', label = 'media')
```

```python
import pyomo_utilities
```

```python
beta = pyomo_utilities.curve_polyfit(df_temp.index, df_temp.loc[:,'Media'], 1)
```

```python
beta
```

```python
yhat1 = beta[0]+beta[1]*(df_temp.index)

plt.figure(figsize=(8,6))
plt.plot(df_temp.index, df_temp.loc[:,'Media'], '*b', label = 'datos')
plt.plot(df_temp.index, yhat1, '-r', label = 'ajuste')
plt.legend(loc = 'best')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid()
plt.show()
```

```python
beta2 = pyomo_utilities.curve_polyfit(df_temp.index, df_temp.loc[:,'Media'], 2)
yhat2 = beta2[0]+beta2[1]*(df_temp.index)+beta2[2]*(df_temp.index)**2

plt.figure(figsize=(8,6))
plt.plot(df_temp.index, df_temp.loc[:,'Media'], '*b', label = 'datos')
plt.plot(df_temp.index, yhat1, '-r', label = 'ajuste 1')
plt.plot(df_temp.index, yhat2, '-g', label = 'ajuste 2')
plt.legend(loc = 'best')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid()
plt.show()
```

```python
beta3 = pyomo_utilities.curve_polyfit(df_temp.index, df_temp.loc[:,'Media'], 3)
yhat3 = beta3[0]+beta3[1]*(df_temp.index)+beta3[2]*(df_temp.index)**2+beta3[3]*(df_temp.index)**3

plt.figure(figsize=(8,6))
plt.plot(df_temp.index, df_temp.loc[:,'Media'], '*b', label = 'datos')
plt.plot(df_temp.index, yhat1, '-r', label = 'ajuste 1')
plt.plot(df_temp.index, yhat2, '-g', label = 'ajuste 2')
plt.plot(df_temp.index, yhat3, '-k', label = 'ajuste 3')
plt.legend(loc = 'best')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid()
plt.show()
```

```python
beta4 = pyomo_utilities.curve_polyfit(df_temp.index, df_temp.loc[:,'Media'], 4)
yhat4 = beta4[0]+beta4[1]*(df_temp.index)+beta4[2]*(df_temp.index)**2+beta4[3]*(df_temp.index)**3+beta4[4]*(df_temp.index)**4

plt.figure(figsize=(8,6))
plt.plot(df_temp.index, df_temp.loc[:,'Media'], '*b', label = 'datos')
plt.plot(df_temp.index, yhat1, '-r', label = 'ajuste 1')
plt.plot(df_temp.index, yhat2, '-g', label = 'ajuste 2')
plt.plot(df_temp.index, yhat3, '-k', label = 'ajuste 3')
plt.plot(df_temp.index, yhat4, '-m', label = 'ajuste 4')
plt.legend(loc = 'best')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid()
plt.show()
```

```python
beta5 = pyomo_utilities.curve_polyfit(df_temp.index, df_temp.loc[:,'Media'], 5)
yhat5 = beta5[0]+beta5[1]*(df_temp.index)+beta5[2]*(df_temp.index)**2+beta5[3]*(df_temp.index)**3+beta5[4]*(df_temp.index)**4+beta5[5]*(df_temp.index)**5

plt.figure(figsize=(8,6))
plt.plot(df_temp.index, df_temp.loc[:,'Media'], '*b', label = 'datos')
plt.plot(df_temp.index, yhat1, '-r', label = 'ajuste 1')
plt.plot(df_temp.index, yhat2, '-g', label = 'ajuste 2')
plt.plot(df_temp.index, yhat3, '-k', label = 'ajuste 3')
plt.plot(df_temp.index, yhat4, '-m', label = 'ajuste 4')
plt.plot(df_temp.index, yhat5, '-r', label = 'ajuste 5')
plt.legend(loc = 'best')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid()
plt.show()
```

```python
beta6 = pyomo_utilities.curve_polyfit(df_temp.index, df_temp.loc[:,'Media'], 6)
yhat6 = beta6[0]+beta6[1]*(df_temp.index)+beta6[2]*(df_temp.index)**2+beta6[3]*(df_temp.index)**3+beta6[4]*(df_temp.index)**4+beta6[5]*(df_temp.index)**5+beta6[6]*(df_temp.index)**6

plt.figure(figsize=(8,6))
plt.plot(df_temp.index, df_temp.loc[:,'Media'], 'b', label = 'datos')
plt.plot(df_temp.index, yhat1, '-r', label = 'ajuste 1')
plt.plot(df_temp.index, yhat2, '-g', label = 'ajuste 2')
plt.plot(df_temp.index, yhat3, '-k', label = 'ajuste 3')
plt.plot(df_temp.index, yhat4, '-m', label = 'ajuste 4')
plt.plot(df_temp.index, yhat5, '-r', label = 'ajuste 5')
plt.plot(df_temp.index, yhat6, '-y', label = 'ajuste 6')
plt.legend(loc = 'best')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid()
plt.show()
```

```python
# np.array([x**i for i in range(21)]).T.dot(beta20_lasso)
beta6_lasso = pyomo_utilities.curve_polyfit(df_temp.index, df_temp.loc[:,'Media'], 6, reg_mode = 'lasso', reg_coef = 0.01)
yhat6 = np.array([(df_temp.index)**i for i in range(7)]).T.dot(beta6_lasso)

plt.figure(figsize=(8,6))
plt.plot(df_temp.index, df_temp.loc[:,'Media'], 'b', label = 'datos')
plt.plot(df_temp.index, yhat1, '-r', label = 'ajuste 1')
plt.plot(df_temp.index, yhat2, '-g', label = 'ajuste 2')
plt.plot(df_temp.index, yhat3, '-k', label = 'ajuste 3')
plt.plot(df_temp.index, yhat4, '-m', label = 'ajuste 4')
plt.plot(df_temp.index, yhat5, '-r', label = 'ajuste 5')
plt.plot(df_temp.index, yhat6, '-y', label = 'ajuste 6 lasso')
plt.legend(loc = 'best')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid()
plt.show()
```
