## Proyecto Simulación Matemática 

Analisis y manejo de datos de Yahoo Finance 

```python
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
tickers = ['^MXX', 'GMEXICOB.MX', 'CM']
data_source = 'yahoo'
start_date = '2011-01-01'
end_date = '2016-12-31'
panel_data = data.DataReader(tickers, data_source, start_date, end_date)
```

```python
ajustes = panel_data.ix['Adj Close']
all_weekdays = pd.date_range(start=start_date, end=end_date, freq='B')

# Reindizamos en este orden
ajustes = ajustes.reindex(all_weekdays)
ajustes = ajustes.fillna(method='ffill')
ajustes
```

```python
ajustes.plot(figsize = (8,10));
```

```python
c = np.cov(ajustes)
c
```
